### Matplotlib 

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt

#creating dataset
data = {'c':20, 'c++':30, 'Java': 50, 'Python': 90}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10,5))

#creating bar plot
plt.bar(courses, values, color = 'maroon', width = 0.4)
plt.xlabel("Courses offered")
plt.ylabel("No of students enrolled")
plt.title("Students enrolled in different cources")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['CSE','ECE','EEE','CIVIL','MECH']
boys = [200,189,100,120,50]
girls = [200,90,30,20,5]
w = 0.6

plt.bar(x, boys,w, label = 'boys')
plt.bar(x, girls,w, bottom = boys, label = 'girls')

#creating bar plot
plt.xlabel("Batches")
plt.ylabel("Students")
plt.title("Students enrolled in different Batches")
plt.legend()
plt.show()
```

```python
import numpy
import matplotlib.pyplot
data = [[30,25,50,20],
        [40,23,51,17],
        [35,22,45,19]]
x = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x + 0.00, data[0], color = 'b', width = 0.25)
ax.bar(x + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(x + 0.50, data[2], color = 'r', width = 0.25)
```

```python
import matplotlib.pyplot
cars = ['Tata','Ferrari','Mercedis','Audi','BMW']
data = [30,34,50,20,40]

#creating plot
fig = plt.figure(figsize=(10,7))
plt.pie(data,labels = cars)

#show plot
plt.show()
```
